```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set style for professional visualizations
sns.set_style('whitegrid')

print("=" * 70)
print("PROJECT 3: SQL DATA ANALYSIS")
print("=" * 70)

# ------------------------------------------------------------------------------
# 1. LOAD DATA AND CREATE DATABASE
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("1. LOADING DATA AND CREATING DATABASE")
print("-" * 50)

# Load the dataset
df = pd.read_excel('Dataset for Data Analytics (3).xlsx')

# Clean the data first (as we did in Project 1)
df['Date'] = pd.to_datetime(df['Date']).dt.date
df['CouponCode'] = df['CouponCode'].fillna('No Code')
text_columns = ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource', 'ShippingAddress', 'CustomerID']
for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.title()
# Corrected: This line should apply directly to 'PaymentMethod' and not be inside the previous loop
df['PaymentMethod'] = df['PaymentMethod'].replace('Online', 'Online Card')
numeric_cols = ['UnitPrice', 'TotalPrice', 'Quantity']
for col in numeric_cols:
    df[col] = df[col].round(2)

print(f"Data loaded: {len(df)} rows, {len(df.columns)} columns")

# Create SQLite database in memory
conn = sqlite3.connect(':memory:')

# Load DataFrame into SQLite table
df.to_sql('orders', conn, index=False, if_exists='replace')
print("Database created with table: orders")

# Verify table exists
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(f"Tables in database: {cursor.fetchall()}")

# ------------------------------------------------------------------------------
# 2. BASIC SELECT AND WHERE CLAUSES
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("2. BASIC SELECT AND WHERE CLAUSES")
print("-" * 50)

# Query 1: Select all columns for first 10 rows
query1 = """
SELECT * FROM orders LIMIT 10;
"""
print("\nQuery 1: First 10 orders")
display(pd.read_sql_query(query1, conn))

# Query 2: Select specific columns
query2 = """
SELECT OrderID, Date, Product, Quantity, TotalPrice
FROM orders
LIMIT 5;
"""
print("\nQuery 2: Selected columns (OrderID, Date, Product, Quantity, TotalPrice)")
display(pd.read_sql_query(query2, conn))

# Query 3: WHERE clause - equality
query3 = """
SELECT OrderID, Product, Quantity, TotalPrice
FROM orders
WHERE Product = 'Monitor'
LIMIT 5;
"""
print("\nQuery 3: WHERE clause (Product = 'Monitor')")
display(pd.read_sql_query(query3, conn))

# Query 4: WHERE clause - comparison
query4 = """
SELECT OrderID, Product, Quantity, TotalPrice
FROM orders
WHERE TotalPrice > 2000
LIMIT 5;
"""
print("\nQuery 4: WHERE clause (TotalPrice > 2000)")
display(pd.read_sql_query(query4, conn))

# Query 5: WHERE clause - pattern matching (LIKE)
query5 = """
SELECT OrderID, CustomerID, ShippingAddress
FROM orders
WHERE ShippingAddress LIKE '%Main St'
LIMIT 5;
"""
print("\nQuery 5: WHERE clause (ShippingAddress LIKE '%Main St')")
display(pd.read_sql_query(query5, conn))

# Query 6: Multiple conditions (AND/OR)
query6 = """
SELECT OrderID, Product, OrderStatus, PaymentMethod
FROM orders
WHERE OrderStatus = 'Delivered' AND PaymentMethod = 'Credit Card'
LIMIT 5;
"""
print("\nQuery 6: WHERE clause (OrderStatus = 'Delivered' AND PaymentMethod = 'Credit Card')")
display(pd.read_sql_query(query6, conn))

# ------------------------------------------------------------------------------
# 3. GROUP BY AND AGGREGATIONS
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("3. GROUP BY AND AGGREGATIONS")
print("-" * 50)

# Query 7: GROUP BY with COUNT
query7 = """
SELECT Product, COUNT(*) AS OrderCount
FROM orders
GROUP BY Product
ORDER BY OrderCount DESC;
"""
print("\nQuery 7: Product order counts (GROUP BY with COUNT)")
display(pd.read_sql_query(query7, conn))

# Query 8: GROUP BY with SUM
query8 = """
SELECT Product, SUM(TotalPrice) AS TotalRevenue
FROM orders
GROUP BY Product
ORDER BY TotalRevenue DESC;
"""
print("\nQuery 8: Total revenue by product (GROUP BY with SUM)")
display(pd.read_sql_query(query8, conn))

# Query 9: GROUP BY with AVG
query9 = """
SELECT Product, AVG(TotalPrice) AS AvgOrderValue
FROM orders
GROUP BY Product
ORDER BY AvgOrderValue DESC;
"""
print("\nQuery 9: Average order value by product (GROUP BY with AVG)")
display(pd.read_sql_query(query9, conn))

# Query 10: GROUP BY with multiple aggregates
query10 = """
SELECT
    Product,
    COUNT(*) AS OrderCount,
    SUM(TotalPrice) AS TotalRevenue,
    AVG(TotalPrice) AS AvgOrderValue,
    MIN(TotalPrice) AS MinOrder,
    MAX(TotalPrice) AS MaxOrder
FROM orders
GROUP BY Product
ORDER BY TotalRevenue DESC;
"""
print("\nQuery 10: Multiple aggregates by product")
display(pd.read_sql_query(query10, conn))

# Query 11: GROUP BY on PaymentMethod
query11 = """
SELECT
    PaymentMethod,
    COUNT(*) AS OrderCount,
    SUM(TotalPrice) AS TotalRevenue,
    AVG(TotalPrice) AS AvgOrderValue
FROM orders
GROUP BY PaymentMethod
ORDER BY TotalRevenue DESC;
"""
print("\nQuery 11: Payment method analysis")
display(pd.read_sql_query(query11, conn))

# Query 12: GROUP BY on OrderStatus
query12 = """
SELECT
    OrderStatus,
    COUNT(*) AS OrderCount,
    SUM(TotalPrice) AS TotalRevenue,
    AVG(TotalPrice) AS AvgOrderValue
FROM orders
GROUP BY OrderStatus
ORDER BY OrderCount DESC;
"""
print("\nQuery 12: Order status analysis")
display(pd.read_sql_query(query12, conn))

# Query 13: GROUP BY on ReferralSource
query13 = """
SELECT
    ReferralSource,
    COUNT(*) AS OrderCount,
    SUM(TotalPrice) AS TotalRevenue,
    AVG(TotalPrice) AS AvgOrderValue
FROM orders
GROUP BY ReferralSource
ORDER BY TotalRevenue DESC;
"""
print("\nQuery 13: Referral source analysis")
display(pd.read_sql_query(query13, conn))

# ------------------------------------------------------------------------------
# 4. HAVING CLAUSE (Filtering Groups)
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("4. HAVING CLAUSE (Filtering Groups)")
print("-" * 50)

# Query 14: HAVING clause
query14 = """
SELECT
    Product,
    COUNT(*) AS OrderCount,
    SUM(TotalPrice) AS TotalRevenue
FROM orders
GROUP BY Product
HAVING COUNT(*) >= 40
ORDER BY TotalRevenue DESC;
"""
print("\nQuery 14: Products with >= 40 orders (HAVING clause)")
display(pd.read_sql_query(query14, conn))

# Query 15: HAVING with AVG
query15 = """
SELECT
    Product,
    AVG(TotalPrice) AS AvgOrderValue,
    COUNT(*) AS OrderCount
FROM orders
GROUP BY Product
HAVING AVG(TotalPrice) > 1500
ORDER BY AvgOrderValue DESC;
"""
print("\nQuery 15: Products with avg order value > $1,500")
display(pd.read_sql_query(query15, conn))

# ------------------------------------------------------------------------------
# 5. ORDER BY CLAUSE
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("5. ORDER BY CLAUSE")
print("-" * 50)

# Query 16: ORDER BY ascending
query16 = """
SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice ASC
LIMIT 5;
"""
print("\nQuery 16: Lowest value orders (ORDER BY ASC)")
display(pd.read_sql_query(query16, conn))

# Query 17: ORDER BY descending
query17 = """
SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 5;
"""
print("\nQuery 17: Highest value orders (ORDER BY DESC)")
display(pd.read_sql_query(query17, conn))

# Query 18: Multiple ORDER BY
query18 = """
SELECT OrderID, Product, Quantity, TotalPrice
FROM orders
ORDER BY Product ASC, TotalPrice DESC
LIMIT 10;
"""
print("\nQuery 18: ORDER BY Product ASC, TotalPrice DESC")
display(pd.read_sql_query(query18, conn))

# ------------------------------------------------------------------------------
# 6. COMPLEX QUERIES (Subqueries, Joins, etc.)
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("6. COMPLEX QUERIES")
print("-" * 50)

# Query 19: Subquery - orders above average
query19 = """
SELECT OrderID, Product, TotalPrice
FROM orders
WHERE TotalPrice > (SELECT AVG(TotalPrice) FROM orders)
ORDER BY TotalPrice DESC
LIMIT 5;
"""
print("\nQuery 19: Orders above average TotalPrice (subquery)")
display(pd.read_sql_query(query19, conn))

# Query 20: Subquery with GROUP BY
query20 = """
SELECT Product, TotalRevenue
FROM (
    SELECT Product, SUM(TotalPrice) AS TotalRevenue
    FROM orders
    GROUP BY Product
) AS product_revenue
WHERE TotalRevenue > 50000
ORDER BY TotalRevenue DESC;
"""
print("\nQuery 20: Products with TotalRevenue > $50,000")
display(pd.read_sql_query(query20, conn))

# Query 21: CASE statements
query21 = """
SELECT
    OrderID,
    TotalPrice,
    CASE
        WHEN TotalPrice < 500 THEN 'Low'
        WHEN TotalPrice BETWEEN 500 AND 2000 THEN 'Medium'
        ELSE 'High'
    END AS OrderTier
FROM orders
LIMIT 10;
"""
print("\nQuery 21: Order tier classification (CASE statement)")
display(pd.read_sql_query(query21, conn))

# ------------------------------------------------------------------------------
# 7. EXECUTION ORDER DEMONSTRATION
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("7. EXECUTION ORDER DEMONSTRATION")
print("-" * 50)

# Show the alias trap - this WILL work in SQLite because of how it processes
# But demonstrates the concept from the PDF
query22 = """
SELECT
    Product,
    SUM(TotalPrice) AS TotalRevenue
FROM orders
GROUP BY Product
HAVING TotalRevenue > 50000
ORDER BY TotalRevenue DESC;
"""
print("\nQuery 22: Using alias in HAVING (demonstrates execution order concept)")
display(pd.read_sql_query(query22, conn))

# ------------------------------------------------------------------------------
# 8. SAVE RESULTS TO EXCEL
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("8. SAVING QUERY RESULTS")
print("-" * 50)

# Collect all query results into a dictionary for export
query_results = {
    'Product_Counts': pd.read_sql_query(query7, conn),
    'Product_Revenue': pd.read_sql_query(query8, conn),
    'Product_Average': pd.read_sql_query(query9, conn),
    'Product_Aggregates': pd.read_sql_query(query10, conn),
    'Payment_Analysis': pd.read_sql_query(query11, conn),
    'OrderStatus_Analysis': pd.read_sql_query(query12, conn),
    'Referral_Analysis': pd.read_sql_query(query13, conn),
    'Products_High_Orders': pd.read_sql_query(query14, conn),
    'High_Avg_Orders': pd.read_sql_query(query15, conn),
    'Top_Orders': pd.read_sql_query(query17, conn),
    'Above_Avg_Orders': pd.read_sql_query(query19, conn),
    'Product_Tiers': pd.read_sql_query(query21, conn),
}

# Save to Excel
with pd.ExcelWriter('Project3_SQL_Results.xlsx', engine='openpyxl') as writer:
    for sheet_name, data in query_results.items():
        data.to_excel(writer, sheet_name=sheet_name, index=False)

print("SQL query results saved to: Project3_SQL_Results.xlsx")

# ------------------------------------------------------------------------------
# 9. ALL SQL QUERIES SUMMARY
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("9. SQL QUERIES SUMMARY")
print("-" * 50)

queries_summary = [
    "1. Basic SELECT - First 10 orders",
    "2. SELECT specific columns",
    "3. WHERE clause - Product = 'Monitor'",
    "4. WHERE clause - TotalPrice > 2000",
    "5. WHERE clause - LIKE pattern matching",
    "6. WHERE clause - AND/OR conditions",
    "7. GROUP BY with COUNT - Product order counts",
    "8. GROUP BY with SUM - Product revenue",
    "9. GROUP BY with AVG - Average order value",
    "10. GROUP BY with multiple aggregates",
    "11. GROUP BY - Payment method analysis",
    "12. GROUP BY - Order status analysis",
    "13. GROUP BY - Referral source analysis",
    "14. HAVING clause - Products with >= 40 orders",
    "15. HAVING clause - Avg order value > $1500",
    "16. ORDER BY ASC - Lowest value orders",
    "17. ORDER BY DESC - Highest value orders",
    "18. Multiple ORDER BY - Product ASC, TotalPrice DESC",
    "19. Subquery - Orders above average",
    "20. Subquery with GROUP BY - Products > $50k",
    "21. CASE statement - Order tier classification",
]

print("\nSQL Queries Executed:")
for i, query_desc in enumerate(queries_summary, 1):
    print(f"  {i}. {query_desc}")

# ------------------------------------------------------------------------------
# 10. VISUALIZATIONS FROM QUERY RESULTS
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("10. VISUALIZATIONS")
print("-" * 50)

# Revenue by Product
product_revenue = pd.read_sql_query(query8, conn)

plt.figure(figsize=(10, 6))
plt.bar(product_revenue['Product'], product_revenue['TotalRevenue'])
plt.title('Total Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sql_product_revenue.png', dpi=300, bbox_inches='tight')
plt.show()

# Order Count by Payment Method
payment_counts = pd.read_sql_query(query11, conn)

plt.figure(figsize=(10, 6))
plt.bar(payment_counts['PaymentMethod'], payment_counts['OrderCount'])
plt.title('Order Count by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sql_payment_counts.png', dpi=300, bbox_inches='tight')
plt.show()

# Order Status Distribution
status_counts = pd.read_sql_query(query12, conn)

plt.figure(figsize=(10, 6))
plt.bar(status_counts['OrderStatus'], status_counts['OrderCount'])
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sql_status_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

# ------------------------------------------------------------------------------
# 11. SUMMARY REPORT
# ------------------------------------------------------------------------------
print("\n" + "-" * 50)
print("11. SUMMARY REPORT")
print("-" * 50)

print(f"""
SQL DATA ANALYSIS COMPLETE
==========================
Database: SQLite (In-Memory)
Table: orders
Total Rows: {len(df)}
Total Columns: 14

Queries Executed: 21
Key SQL Concepts Demonstrated:
  - SELECT statements
  - WHERE clause (equality, comparison, pattern matching)
  - GROUP BY clause
  - Aggregate functions (COUNT, SUM, AVG, MIN, MAX)
  - HAVING clause
  - ORDER BY clause
  - Subqueries
  - CASE statements

Output Files:
  - Project3_SQL_Results.xlsx (12 sheets with query results)
  - sql_product_revenue.png
  - sql_payment_counts.png
  - sql_status_distribution.png
""")

# Close connection
conn.close()
print("Database connection closed.")
print("\n" + "=" * 70)
print("PROJECT 3: SQL DATA ANALYSIS COMPLETE")
print("=" * 70)
```

### Boxplot of TotalPrice by Product

This visualization will show the distribution of `TotalPrice` for each `Product`. Boxplots are excellent for identifying:
*   **Median:** The line in the middle of the box.
*   **Interquartile Range (IQR):** The box itself, representing the middle 50% of the data.
*   **Whiskers:** The lines extending from the box, indicating the range of the data (usually 1.5 times the IQR).
*   **Outliers:** Points beyond the whiskers.


```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Product', y='TotalPrice', data=df)
plt.title('Distribution of TotalPrice by Product')
plt.xlabel('Product')
plt.ylabel('TotalPrice ($)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('sql_totalprice_by_product_boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```
